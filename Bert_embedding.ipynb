```python
from transformers import AutoTokenizer, AutoModel #導入庫
model=AutoModel.from_pretrained("bert-base-uncased"); ## 載入 BERT 模型
print(model) # 打印模型結構
```

```python
print(model.num_parameters()) ## 打印模型的參數數量
```

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased") # 載入預訓練的 BERT 分詞器
raw_inputs = [ # 定義原始輸入文本
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt") # 將文本轉換為模型可接受的格式
print(inputs) # 打印處理後的輸入
```
